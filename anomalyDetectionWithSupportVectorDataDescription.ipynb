```python
import swat
import pandas as pd
import numpy as np
from _config import login
import os
```

```python
os.environ['CAS_CLIENT_SSL_CA_LIST']=r'/ext_str/beast_trustfiles/nBeast_trustfiles/trustedcerts.pem'
host = host = login()[2]
user,pswd = login()[0],login()[1]
conn = swat.CAS(host,5570,user,pswd)
```

```python
df = pd.read_csv("/ext_str/data/pred_maintenance_data/creditcard.csv")
df.head()
```

```python
# Convert Types
for col in ['Time', 'Class']:
    df[col] = df[col].astype('object')
```

```python
conn.upload(df,casout=dict(caslib="Public",name="sensor_data_orig"))  #Upload to CAS
```

```python
conn.setsessopt(caslib="Public")
sensor_df = conn.CASTable("sensor_data_orig")
sensor_df.head()
```

```python
# Drop the unnecessary column
sensor_df.altertable(columns= [dict(name="Amount",drop=True)])
sensor_df.columninfo()
```

```python
conn.loadactionset(actionset='svdd')
conn.loadactionset(actionset='fedSQL')
```

```python
conn.datastep.runcode(
code = """data public.sensor_train public.sensor_score;
set public.sensor_data_orig;
if class eq 0 then output public.sensor_train;
else output public.sensor_score;
run;"""
)
```

```python
train_df = conn.CASTable("sensor_train")
train_df.altertable(columns= [dict(name="Class",drop=True)])
train_df.columninfo()
```

```python
train_df.head()
```

```python
feature_list = [i for i in train_df if 'V' in i]
# model spec - could possibly use tuneMethod param 
conn.svdd.svddtrain(table=train_df,inputs=feature_list,bw=2.8290,solver="ACTSET", 
                    savestate="sense_model",
                   output=dict(casout=dict(name="train_model",replace=True)),
                    id = ["Time"])
```

```python
conn.tableinfo()
```

```python
outmodel = conn.CASTable("sense_model")
#outmodel.head()
conn.loadactionset(actionset='astore')
```

```python
# Score the score data and get the Summary by Group 
score_df = conn.CASTable("sensor_score")
score_df.altertable(columns= [dict(name="Class",drop=True)])
result = conn.aStore.score(out = {"name":"output_df","replace":True}
                      ,rstore = "sense_model"
                      ,table=score_df
                     )
out = conn.CASTable("output_df")
out.groupby("_SVDDSCORE_").summary()
```

```python
test = conn.aStore.score(out = {"name":"sensor_df_output","replace":True}
                      ,rstore = "sense_model"
                      ,table=sensor_df
                     )
```

```python
sensor_out = conn.CASTable('sensor_df_output')
sensor_out.groupby("_SVDDSCORE_").summary()
```
